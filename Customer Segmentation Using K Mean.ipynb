```python
#Keep only United Kingdom data
usdata = data.query("Country=='United Kingdom'").reset_index(drop=True)
usdata.head()
```

```python
usdata.isnull().sum()
```

```python
#remove missing value in CustomerID
usdata = usdata[pd.notnull(usdata['CustomerID'])]
```

```python
#Validate if there are any negative values in Quantity column
usdata.Quantity.min()
```

```python
#Validate if there are any negative values in UnitPrice column
usdata.UnitPrice.min()
```

```python
usdata=usdata[(usdata["Quantity"]>0)]
usdata.head()
```

```python
#Convert the string date field to datetime
usdata['InvoiceDate'] = pd.to_datetime(usdata['InvoiceDate'])
usdata.head()
```

```python
usdata["TotalAmount"]=usdata["Quantity"]*usdata["UnitPrice"]
usdata.head()
```

```python
import datetime as dt
latestdate=dt.datetime(2011,12,10)
x=usdata.groupby("CustomerID").agg({'InvoiceDate':lambda x:(latestdate-x.max()).days,
                                    'InvoiceNo':lambda x:len(x),
                                   'TotalAmount':lambda x:x.sum()})
x.head()
```

```python
x.info()
```

```python
x.rename(columns={'InvoiceDate':'Recency','InvoiceNo':'Frequecy','TotalAmount':'Monetry'},inplace=True)
x.head()
```

```python
x.reset_index().head()
```

```python
quantiles=x.quantile(q=[0.25,0.5,0.75])
quantiles.to_dict()
```

```python
def rscore(a,b,c):
    if a<=c[b][0.25]:
        return 1
    elif a<=c[b][0.50]:
        return 2
    elif a<=c[b][0.75]:
        return 3
    else:return 4
    
def fmscore(a,b,c):
    if a<=c[b][0.25]:
        return 4
    elif a<=c[b][0.50]:
        return 3
    elif a<=c[b][0.75]:
        return 2
    else:return 1
```

```python
x['R'] = x['Recency'].apply(rscore, args=('Recency',quantiles,))
x['F'] = x['Frequecy'].apply(fmscore, args=('Frequecy',quantiles,))
x['M'] = x['Monetry'].apply(fmscore, args=('Monetry',quantiles,))
x.head()
```

```python
x["RFM_Score"]=x["R"]+x["F"]+x["M"]
x.head()
```

```python
loyality_label=['Platinum','Gold','Silver','Bronze']
x["Loyality_Score"]=pd.qcut(x.RFM_Score,q=4,labels=loyality_label)
x.reset_index().head()
```

```python
(x==0).sum()
```

```python
def handle_neg_n_zero(num):
    if num <= 0:
        return 1
    else:
        return num
#Apply handle_neg_n_zero function to Recency and Monetary columns 
x['Recency'] = [handle_neg_n_zero(x) for x in x.Recency]
x['Monetary'] = [handle_neg_n_zero(x) for x in x.Monetry]

#Perform Log transformation to bring data into normal or near normal distribution
logData = x[['Recency', 'Frequecy', 'Monetary']].apply(np.log, axis = 1).round(3)
logData.head()
```
